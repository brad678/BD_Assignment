````python
import os
import feedparser, json
from langchain.chains import SequentialChain
from langchain_openai import AzureChatOpenAI
from langchain.schema.output_parser import StrOutputParser
from langchain.prompts import PromptTemplate
from langchain.schema.runnable import RunnableLambda
from langfuse import Langfuse
from langfuse.callback import CallbackHandler


# âœ… Set Langfuse env vars BEFORE using Langfuse
os.environ['LANGFUSE_SECRET_KEY']="sk-lf-1a8132d8-bbab-4983-8a99-f8b4f88c93fc"
os.environ['LANGFUSE_PUBLIC_KEY']="pk-lf-ef0e83e1-087d-43ed-994a-a73ffddc0021"
os.environ['LANGFUSE_HOST']="https://cloud.langfuse.com"

# âœ… Initialize Langfuse and CallbackHandler
# langfuse = Langfuse()
langfuse_handler = CallbackHandler()

class StockNewsLangChain:
    def __init__(self, azure_endpoint, api_key, deployment_name, company_name):
        self.azure_endpoint = azure_endpoint
        self.api_key = api_key
        self.deployment_name = deployment_name
        self.company_name = company_name

        # âœ… Pass Langfuse callback to LLM
        self.llm = AzureChatOpenAI(
            azure_endpoint=self.azure_endpoint,
            api_key=self.api_key,
            deployment_name=self.deployment_name,
            api_version="2024-12-01-preview"
        )

    def fetch_news(self, stock_code):
        url = f"https://news.google.com/rss/search?q={stock_code}+stock"
        feed = feedparser.parse(url)
        return [entry.title for entry in feed.entries]

    def run_chain(self):
        prompt_stock_code = PromptTemplate.from_template(
            "Provide only the four-character stock ticker symbol for {company_name}. No additional details."
        )
        chain_stock_code = prompt_stock_code | self.llm | StrOutputParser()

        fetch_news_lambda = RunnableLambda(self.fetch_news)

        prompt_stock_news = PromptTemplate.from_template(
            "Summarize these financial news headlines:\n{headlines}"
        )
        chain_summarization = prompt_stock_news | self.llm | StrOutputParser()

        stock_code = chain_stock_code.invoke({"company_name": self.company_name}, 
                                             config={"callbacks":[langfuse_handler]})
        news_headlines = self.fetch_news(stock_code)
        news_summary = chain_summarization.invoke({"headlines": news_headlines}, 
                                                  config={"callbacks":[langfuse_handler]})

        return {
            "company_name": self.company_name,
            "stock_code": stock_code,
            "news_headlines": news_headlines,
            "summary": news_summary
        }

# Example usage
AZURE_ENDPOINT = os.getenv('AZURE_OPENAI_ENDPOINT')
API_KEY = os.getenv('AZURE_OPENAI_API_KEY')
DEPLOYMENT_NAME = os.getenv('AZURE_OPENAI_DEPLOYMENT')
COMPANY_NAME = "Tesla"

stock_news_chain = StockNewsLangChain(AZURE_ENDPOINT, API_KEY, DEPLOYMENT_NAME, COMPANY_NAME)
output = stock_news_chain.run_chain()
print(json.dumps(output, indent=4))
````

````python


class StockNewsSentimentAnalyzer:
    """Classifies sentiment, extracts named entities, and assesses market impact using GPT-4o Mini."""

    def __init__(self, azure_endpoint, api_key, deployment_name, news_summary, news_company_name, news_stock_code, news_desc):
        self.azure_endpoint = azure_endpoint
        self.api_key = api_key
        self.deployment_name = deployment_name
        self.news_summary = news_summary
        self.news_company_name = news_company_name
        self.news_stock_code = news_stock_code
        self.news_desc = news_desc

        # Initialize Azure OpenAI client
        self.llm = AzureChatOpenAI(
            azure_endpoint=self.azure_endpoint,
            api_key=self.api_key,
            deployment_name=self.deployment_name,
            api_version="2024-12-01-preview"
        )

    def analyze_sentiment_and_entities(self):
        """Classify sentiment, extract named entities, and determine market implications."""

        prompt = PromptTemplate.from_template("""
        Given the following financial news summary, perform the following tasks:
        
        1. **Classify sentiment** as 'Positive', 'Negative', or 'Neutral'.
        2. **Extract named entities**, including:
            - **People** (individuals mentioned)
            - **Places** (locations related to the news)
            - **Companies** (primary company and other firms referenced)
        3. **Identify related industries** relevant to the news.
        4. **Determine market implications**, such as:
            - Stock trends (bullish, bearish, volatile)
            - Economic impacts
            - Investor sentiment changes
        5. **Assign a confidence score** (range: 0-1) based on certainty in the analysis.

        **News Summary**:
        {news_summary}

        Provide the output JSON strictly in the following format:
        {{
            "company_name": "{{news_company_name}}",
            "stock_code": "{{news_stock_code}}",
            "newsdesc": "{{news_desc}}",                                  
            "sentiment": "<Positive/Negative/Neutral>",
            "named_entities": {{
                "people": [List of names],
                "places": [List of locations],
                "companies": {{
                    "primary_company": "<Main company>",
                    "other_companies": [List of referenced companies]
                }},
            }},
            "related_industries": [List of industries],
            "market_implications": {{
                "stock_trend": "<Bullish/Bearish/Volatile>",
                "economic_impact": "<Brief description>",
                "investor_sentiment": "<Brief description>"
            }},
            "confidence_score": <float between 0 and 1>
        }}
        """)

        chain = prompt | self.llm | StrOutputParser()
        output_str = chain.invoke({"news_summary": self.news_summary},
                                  config={"callbacks":[langfuse_handler]})

        def clean_json_output(output_str):
            """Removes Markdown-style triple backticks from JSON string."""
            if output_str.startswith("```json"):
                output_str = output_str.lstrip("```json").rstrip("```")  # Remove formatting markers
            return output_str.strip()

        # Apply cleanup before parsing
        try:
            cleaned_output = clean_json_output(output_str)
            structured_output = json.loads(cleaned_output)  # Convert to proper JSON
        except json.JSONDecodeError as jde:
            print(f"JSON Decode Error: {jde}")
            structured_output = None

        return structured_output


# ðŸ”¥ **Example Usage**
news_summary = output['summary']  # Use the summary from the previous step
news_company_name = output['company_name']
news_stock_code = output['stock_code']
news_desc = output['news_headlines']

# Instantiate the class
sentiment_analyzer = StockNewsSentimentAnalyzer(AZURE_ENDPOINT, API_KEY, DEPLOYMENT_NAME, news_summary, news_company_name, news_stock_code, news_desc)

# Get structured output
structured_output = sentiment_analyzer.analyze_sentiment_and_entities()

print(json.dumps(structured_output, indent=4))
````
